```python
import cv2
import glob
import os

# === Параметри ===
image_folder = './images/sequence-2'                # папката с изображенията
image_type = '*.jpg'                   # или '*.jpg', '*.jpeg', '*.bmp'
output_path = 'video/video_1.mp4'
fps = 1                               # кадри в секунда

# === Събиране на изображения ===
image_paths = sorted(glob.glob(os.path.join(image_folder, image_type)))

# === Проверка за поне едно изображение ===
if not image_paths:
    raise ValueError("Няма изображения в зададената папка.")

# === Прочитане на първото изображение, за да вземем размерите ===
frame = cv2.imread(image_paths[0])
height, width = frame.shape[:2]

# === Инициализиране на видео запис ===
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# === Добавяне на изображения ===
for path in image_paths:
    img = cv2.imread(path)

    # по желание: преоразмеряване до еднакъв размер:
    img = cv2.resize(img, (width, height))

    print(path)
    out.write(img)

# === Завършване ===
out.release()
print(f"Видео записано като {output_path}")
```
